#### Classify movies based on ratings(Excellent, Good & Average)

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['Rating_Ctegory']=data['Rating'].apply(rating)
```

#### find unique values from Genre

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_genre=[]
for item in list1:
    for gen in item:
        one_genre.append(gen)
```

```python
one_genre
```

```python
unique_genre=[]
for gen in one_genre:
    if gen not in unique_genre:
        unique_genre.append(gen)
```
